```python
df['Storage '] = df['Storage '].str.replace(' ', '') #Menghapus spasi
df['Storage '] = df['Storage '].str.replace('GB','') #Mengubah GB
df['Storage '] = df['Storage '].astype(int) #Mengubah dari string ke int
df['Storage ']
```

Analisis Kolom RAM

```python
df['RAM '].unique()
```

```python
df['RAM '] = df['RAM '].str.replace(' ', '') #Menghapus spasi
df['RAM '] = df['RAM '].str.replace('GB','') #Mengubah GB
df['RAM '] = df['RAM '].astype(int) #Mengubah dari string ke int
df['RAM ']
```

Analisis Kolom Harga / Price

```python
df['Price ($)'].unique()
```

```python
df['Price ($)'] = df['Price ($)'].str.replace('$', '') #Menghapus dollar
df['Price ($)'] = df['Price ($)'].str.replace(',','') #Mengubah koma
df['Price ($)'] = df['Price ($)'].str.replace(' ','') #Menghapus spasi


df['Price ($)'] = df['Price ($)'].astype(int) #Mengubah dari string ke int
df['Price ($)']
```

##Grouping

```python
#Menghitung Rata-Rata Harga Setiap Brand
df.groupby('Brand').agg(rata2_harga = ('Price ($)', 'mean')).sort_values(by = 'rata2_harga')
```

```python
df.groupby('Brand').agg(rata2_harga = ('Price ($)', 'mean'),
                        rata2_battery_dollar = ('Battery Dollar', 'mean')).\
                        sort_values(by = 'rata2_battery_dollar', ascending = False)
```

##Lambda

```python
df.describe()
```

```python
#Membuat satu kolom baru yang berisi kategori Harga HP
#'MURAH' --> Harga < 500 USD, 'MAHAL' --> Harga lainnya
#Dengan Fungsi Lambda
df['Category'] = df['Price ($)'].apply(lambda x: 'MURAH' if x < 500 else 'MAHAL')
```
